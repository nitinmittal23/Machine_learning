```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
dfx = pd.read_csv('weightedX.csv')
dfy = pd.read_csv('weightedY.csv')
x = dfx.values
y = dfy.values
x = x.reshape((-1,))
y = y.reshape((-1,))
X = (x-x.mean())/x.std()
Y = y
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
x1 = X
x2 = X**2
```

```python
X = np.stack((x1,x2),axis =1 )
print(X.shape)
```

```python
model = LinearRegression(normalize=True)
model.fit(X,Y)
output = model.predict(X)
```

```python
plt.scatter(X[:,0],Y,label="data")
plt.scatter(X[:,0],output,color='red',label = 'prediction')
plt.legend()
plt.show()
```
