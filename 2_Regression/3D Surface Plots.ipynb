```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(-4,4,0.01)
b = a
a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2,cmap='coolwarm')
plt.show()
```
