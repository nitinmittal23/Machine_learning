```python
titles = df.get("movie_title")
#print(titles)
```

```python
freq_title = {}
```

```python
for t in titles:
    length = len(t)
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1
```

```python
print(freq_title)
```

```python
x = list(freq_title.keys())
y = list(freq_title.values())
plt.scatter(x,y)
plt.xlabel("length of movie title")
plt.ylabel("freq ")
plt.show()
```

# Normalising

```python
dfx = pd.read_csv("linearX.csv")
dfy = pd.read_csv("linearY.csv")
```

```python
x= dfx.values
y=dfy.values

x2 = (x-x.mean())/x.std()
```

```python
plt.scatter(x,y, label="raw")
plt.legend()
plt.scatter(x2,y,label="normalised")
plt.legend()
```
