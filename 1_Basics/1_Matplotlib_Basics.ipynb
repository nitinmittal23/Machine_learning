### Data Visualisation


```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(10)
x1 = x**2
x2 = x**3
print(x)
```

```python
plt.plot(x,x1,label='2018')
plt.plot(x,x2, color = 'red',label ='2019')
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("Stock price prediction")
plt.show()
```

```python
plt.scatter(x,x*2,color='green',marker = '^',label = 'stock Price')
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("Stock price prediction")
plt.show()
```

# Bar Chart

```python
numbers = np.random.randint(0,10,5)
numbers2 = np.random.randint(0,10,5)
indices = np.arange(5)
print(numbers)
print(indices)
```

```python
plt.bar(indices,numbers,0.25, color="green",label='green')
plt.bar(indices+0.25,numbers2,0.25, color="orange",label='Orange')
plt.legend()
plt.show()
```

## Pie Chart

```python
labels = ['algebra','calculus','coding','algorithms']
values = [90,80,54,40]
plt.pie(values,labels=labels, radius = 1,explode=(0,1,0,0),shadow = True)
plt.show()
```

```python
# generate a normal distribution 1-D
u = 5
sigma = 2

# in standard normal distribution u=1 and sigma = 0 
vals = np.random.randn(100)*sigma + u
print(vals.shape)

plt.hist(vals,10)
plt.show()
```

```python
vals = np.round(vals)
z = np.unique(vals, return_counts=True)
print(z)
```

```python
x = vals
y= np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

# Multivariate normal distribution

```python
# Bivariate
mean = np.array([2,4])
cov = np.array([[1,0.8],[0.8,1]])

mean2 = np.array([3,3])
cov2 = np.array([[1,-1],[-1,2]])

dist = np.random.multivariate_normal(mean,cov,500)

dist2 = np.random.multivariate_normal(mean2,cov2,500)


plt.figure(figsize=(8,8))
plt.subplot(221)
plt.scatter(dist[:,0],dist[:,1],color="red")
plt.subplot(222)
plt.scatter(dist2[:,0],dist2[:,1],color="green")

plt.show()
```
