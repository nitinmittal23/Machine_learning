## pip install -U scikit-learn

```python
from sklearn.datasets import make_regression, make_classification, make_blobs, make_moons
```

```python
from matplotlib import pyplot as plt
```

```python
x,y = make_regression(n_samples =100, n_features=1,n_informative=1, noise=12.8)
```

```python
print(x.shape,y.shape)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
xc,yc = make_classification(n_samples=400,n_features=2,n_classes=3,n_clusters_per_class=1,n_informative=2,n_redundant=0)
```

```python
print(xc.shape,yc.shape)
```

```python
plt.scatter(xc[:,0],xc[:,1], c=yc)
plt.show()
```

# Blob datasets
#### for unsupervised learning

```python
xb,yb =make_blobs(n_samples=100,n_features=2,centers=5)
```

```python
plt.scatter(xb[:,0],xb[:,1],c=yb)
plt.show()
```

## moons data set

```python
X,Y= make_moons(n_samples=100,noise=0.1, random_state=5)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
```

# Subplots


```python
plt.figure(figsize=(8,8))
plt.subplot(221)
plt.title("regression")
plt.scatter(x,y)

plt.subplot(222)
plt.title("classification")
plt.scatter(xc[:,0],xc[:,1],c=yc)

plt.subplot(223)
plt.title("blobs")
plt.scatter(xb[:,0],xb[:,1],c=yb)

plt.subplot(224)
plt.title("moons")
plt.scatter(X[:,0],X[:,1],c=Y)


plt.show()
```
