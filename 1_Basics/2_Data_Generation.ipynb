# Data for Classification
- let us generate some examples for binary Classification

```python
u1 = np.array([2,4])
cov_1 = np.array([[2,0.8],[0.8,1]])
u2 = np.array([-1,-3])
cov_2 = np.array([[1,0.5],[0.5,2]])
```

```python
x1 = np.random.multivariate_normal(u1, cov_1, 300)
print(x1.shape)
x2 = np.random.multivariate_normal(u2, cov_2, 400)
print(x2.shape)
```

```python
plt.scatter(x1[:,0],x1[:,1])
plt.scatter(x2[:,0],x2[:,1], color = "red")
plt.xlabel("x1 size")
plt.ylabel("x2 sweetness")
plt.show()
```

```python
y = np.zeros(700)
y[:300]=1
print(y.shape)
```

```python
x = np.vstack((x1,x2))
print(x.shape)
```

```python
#plt.scatter(x[:,0],x[:,1])
for i in range (700):
    if y[i]==1:
        plt.scatter(x[i,0],x[i,1],color ='green')
    else:
        plt.scatter(x[i,0],x[i,1],color ='red')
plt.show()
```

##Shortcut


```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```
